#Transform refund data

```python
df = spark.table('gizmobox.bronze.py_refunds')
display(df)
```

##extract specific portion of data

```python
from pyspark.sql import functions as F
df1 = df.select(
    df.refund_id,
    df.payment_id,
    F.date_format(df.refund_timestamp,'yyyy-MM-dd').cast('date').alias('refund_date'),
    F.date_format(df.refund_timestamp,'HH:mm:ss').cast('timestamp').alias('refund_time'),
    df.refund_amount,
    F.split(df.refund_reason,':')[0].alias('refund_reason'),
    F.split(df.refund_reason,':')[1].alias('refund_source')
)
display(df1)
```

```python
df1.writeTo('gizmobox.silver.py_refunds').createOrReplace()
```
