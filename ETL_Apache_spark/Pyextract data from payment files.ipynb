##create external tables
We have to specify schema

```python
py_schema = 'payment_id int, order_id int, payment_timestamp timestamp, payment_status int, payment_method string'
```

```python
df = spark.read.format('csv').option('delimeter',(',')).schema(py_schema).load('abfss://gizmobox@databrickslearningmanoj.dfs.core.windows.net/Landing/external_data/payments')
display(df)
```

```python
df.writeTo("gizmobox.bronze.py_payments").createOrReplace()
```
