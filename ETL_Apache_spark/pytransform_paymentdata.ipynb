#Transform Payments data

```python
df = spark.table("gizmobox.bronze.py_payments")
display(df)
```

```python
from pyspark.sql import functions as F
df1 = df.select('payment_id', 
                'order_id',
                F.date_format('payment_timestamp','yyyy-MM-dd').cast('date').alias('payment_date'),
                F.date_format('payment_timestamp','HH:mm:ss').cast('timestamp').alias('payment_time'),
                'payment_status',
                'payment_method'
                )
display(df1)
```

```python
df2 = df1.select("payment_id", 
                 "order_id", 
                 "payment_date", 
                 "payment_time", 
                 F.when(df1.payment_status ==1, 'Success')
                 .when(df1.payment_status ==2, 'Pending')
                 .when(df1.payment_status ==3, 'Cancelled')
                 .when(df1.payment_status ==4, 'Refund').alias('payment_status'), 
                 "payment_method")
display(df2)
```

```python
df2.writeTo('gizmobox.silver.py_payments').createOrReplace()
```
