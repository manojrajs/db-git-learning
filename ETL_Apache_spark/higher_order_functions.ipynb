#Higher Order Functions

Commonly used higher order functions
1. Transform
2. Filter
3. Exists
4. Agggregate

```sql
create or replace temporary view order_items as
select * from
values
(1,array('smarphone','laptop','monitor')),
(2,array('headphones','tablet','smartwatch')),
(3,array('keyboard','mouse'))
AS orders(order_id, items)
```

```sql
select * from order_items
```

##transform

```sql
select order_id, transform(items, x -> upper(x))
from order_items
```

##Filter

```sql
select order_id, Filter(items, x -> x like '%smart%')
from order_items
```

##Exists

```sql
select order_id, Exists(items, x -> x = 'monitor')
from order_items
```

##Array with more than 1 object


```sql
create or replace temporary view order_items as
select * from values
(1,array(
        named_struct('smarphone','electronics', 699),
        named_struct('laptop','electronics', 199),
        named_struct('monitor','electronics', 399)
    )),
    (2,array(
        named_struct('headphones','electronics', 599),
        named_struct('tablet','electronics', 199),
        named_struct('smartwatch','electronics', 299)
    )),
    (3,array(
        named_struct('keyboard','electronics', 89),
        named_struct('mouse','electronics', 59)
    )) AS orders(order_id, items 
    )
```
