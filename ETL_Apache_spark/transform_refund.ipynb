#Transform refund data

```sql
select * from asql_gizmobox_db_conn_catalog.dbo.refunds
```

##extract specific portion of data

```sql
select 
   refund_id,
   payment_id,
   CAST(date_format(refund_timestamp,'yyyy-MM-dd') as date) as refund_date,
   date_format(refund_timestamp,'HH:mm:ss') as refund_time,
   refund_amount,
   split(refund_reason,':')[0] as refund_reason,
   split(refund_reason,':')[1] as refund_source
from asql_gizmobox_db_conn_catalog.dbo.refunds
```

```sql
create table gizmobox.silver.refunds
as
select 
   refund_id,
   payment_id,
   CAST(date_format(refund_timestamp,'yyyy-MM-dd') as date) as refund_date,
   date_format(refund_timestamp,'HH:mm:ss') as refund_time,
   refund_amount,
   split(refund_reason,':')[0] as refund_reason,
   split(refund_reason,':')[1] as refund_source
from asql_gizmobox_db_conn_catalog.dbo.refunds
```
