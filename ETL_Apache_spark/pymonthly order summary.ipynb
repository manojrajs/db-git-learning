#Monthly order summary

```python
df = spark.table('gizmobox.silver.py_orders')
display(df)
```

```python
from pyspark.sql import functions as F
df1 = df.select(
        F.date_format(F.col("transaction_timestamp"), "yyyy-MM").alias("transaction_date"),
        F.col("order_id"),
        F.col("quantity"),
        F.col("price"),
        F.col("customer_id")
    ) \
    .groupBy("transaction_date", "customer_id") \
    .agg(
        F.countDistinct("order_id").alias("total_orders"),
        F.sum("quantity").alias("total_items_bought"),
        F.sum(F.col("price") * F.col("quantity")).alias("total_amount")
    )
display(df1)
```

```python
from pyspark.sql import functions as F
df1 = df.select(
    F.date_format(df.transaction_timestamp, 'yyyy-MM').alias("transaction_date"),
    F.col("order_id"),
    F.col("quantity"),
    F.col("price"),
    F.col("customer_id")
)\
.groupBy('transaction_date','customer_id')\
.agg(
        F.countDistinct("order_id").alias("total_orders"),
        F.sum("quantity").alias("total_items_bought"),
        F.sum(F.col("price") * F.col("quantity")).alias("total_amount")
)
display(df1)
```

```python
df1.writeTo('gizmobox.gold.order_summary').createOrReplace()
```
