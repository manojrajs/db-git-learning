##External location creation

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databrickslearningmanoj_gizmobox
    URL 'abfss://gizmobox@databrickslearningmanoj.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL databricks_ext_strg)
    COMMENT 'Gizmobox'
```

##Catalogue objects Creation

```sql
CREATE CATALOG IF NOT EXISTS Gizmobox
MANAGED LOCATION 'abfss://gizmobox@databrickslearningmanoj.dfs.core.windows.net/'
```

```sql
SELECT current_catalog()
```

```sql
USE CATALOG Gizmobox
```

```sql
CREATE SCHEMA IF NOT EXISTS Gizmobox.Landing
    MANAGED LOCATION 'abfss://gizmobox@databrickslearningmanoj.dfs.core.windows.net/Landing';
CREATE SCHEMA IF NOT EXISTS Gizmobox.Bronze
    MANAGED LOCATION 'abfss://gizmobox@databrickslearningmanoj.dfs.core.windows.net/Bronze';
CREATE SCHEMA IF NOT EXISTS Gizmobox.Gold
    MANAGED LOCATION 'abfss://gizmobox@databrickslearningmanoj.dfs.core.windows.net/Gold';
CREATE SCHEMA IF NOT EXISTS Gizmobox.Silver
    MANAGED LOCATION 'abfss://gizmobox@databrickslearningmanoj.dfs.core.windows.net/Silver';
```

## Create Volume
