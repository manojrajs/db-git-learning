```sql
create or replace function gizmobox.gold.get_full_name(first_name string, surname string)

Returns string
Return concat(initcap(first_name),' ',initcap(surname))
```

```sql
select gizmobox.gold.get_full_name('john','doe')
```

```sql
describe function extended gizmobox.gold.get_full_name 
```

```sql
create or replace function gizmobox.gold.get_payment_status(payment_status string)

Returns string
Return CASE payment_status
        WHEN  1 THEN 'Success' 
        WHEN  2 THEN 'Pending' 
        WHEN  3 THEN 'Cancelled'
        WHEN  4 THEN 'Failed'
        END 
```

```sql
select gizmobox.gold.get_payment_status('1')
```
