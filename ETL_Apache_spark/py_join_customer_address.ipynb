#join customer and address

```python
df_customers = spark.table('gizmobox.silver.py_customers')
display(df_customers)
```

```python
df_address = spark.table('gizmobox.silver.py_address')
display(df_address)
```

```python
df_join = df_customers.join(df_address,(df_customers.customer_id == df_address.customer_id), "inner").select(
    "customer_name",
    "date_of_birth",
    "email",
    "telephone",
    "member_since",
    "shipping_address_line_1",
    "shipping_city",
    "shipping_state",
    "shipping_postcode",
    "billing_address_line_1",
    "billing_city",
    "billing_state",
    "billing_postcode"
)
display(df_join)
```

```python
df_join.writeTo('gizmobox.gold.py_customer_address').createOrReplace()
```
