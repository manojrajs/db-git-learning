#Transform membership data

```sql
select * from gizmobox.bronze.v_membership
```

```sql
select 
    path,
    regexp_extract(path,'.*/([0-9]+)\\.png$',1) as customer_id,
    modificationTime,
    length,
    content as membership_card
from gizmobox.bronze.v_membership
```

```sql
create table gizmobox.silver.memberships
as
select 
    regexp_extract(path,'.*/([0-9]+)\\.png$',1) as customer_id,
    content as membership_card
from gizmobox.bronze.v_membership
```

```sql
select * from gizmobox.silver.memberships
```
