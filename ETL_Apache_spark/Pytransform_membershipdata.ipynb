#Transform membership data

```python
df = spark.table('gizmobox.bronze.v_membership')
display(df)
```

```python
from pyspark.sql import functions as F
df1 = df.select(
    F.regexp_extract(df.path,r".*/([0-9]+)\.png$",1).alias("customer_id"),
    # F.regexp_extract(F.col("path"), r"([0-9]+)\.png$", 1).alias("customer_id"),
    F.col("content").alias("membership_card")
)
display(df1)
```

```python
df1.writeTo("gizmobox.silver.py_membership").createOrReplace()
```
