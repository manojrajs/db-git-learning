#Transform Address Data

Pivot Clause

```python
df =  spark.table('gizmobox.bronze.v_address')
display(df)
```

```python
from pyspark.sql import functions as F
df1 = (
    df
    .groupBy('customer_id')
    .pivot('address_type',['billing','shipping'])
    .agg(
        F.max('address_line_1').alias('address_line_1'),
        F.max('city').alias('city'),
        F.max('state').alias('state'),
        F.max('postcode').alias('postcode')
        )
)  
display(df1)
```

```python
df1.writeTo('Gizmobox.silver.py_address').createOrReplace()
```
