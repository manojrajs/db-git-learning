##Extract data using Pyspark

```python
df = spark.sql('select * from json.`/Volumes/gizmobox/landing/operational_data/customers/`')
display(df)
```

```python
df = spark.sql('create or replace temporary view tv_customers as select * from json.`/Volumes/gizmobox/landing/operational_data/customers/`')
display(df)
```

```python
%sql select * from tv_customers
```

##spark dataframe reader API

```python
df = spark.read.format('json').load('/Volumes/gizmobox/landing/operational_data/customers/')
display(df)
```
