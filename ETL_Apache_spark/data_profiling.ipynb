##data profiling using code

```python
%sql
select count(*), count(customer_id) from gizmobox.bronze.v_customer
```

```python
%sql
select count(*), count_if(customer_id is null) from gizmobox.bronze.v_customer
```

```python
%sql
select count(*),count(distinct customer_id) from gizmobox.bronze.v_customer
where customer_id is not null
```
