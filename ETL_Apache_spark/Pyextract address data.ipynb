##extract address data
We cant extract using CSV (has header and delimeter is ',')

```python
df = spark.read.format('csv').option('header', 'true').option('delimiter', '\t').load('/Volumes/gizmobox/landing/operational_data/addresses/')
display(df)
```

```python
df.writeTo('gizmobox.bronze.py_address').createOrReplace()
```
